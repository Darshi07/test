## program to find mean median and mode of a give list of numbers

```python
list = [1,2,2,3,4,5,6,2,2]
s=0

for i in range(len(list)):
    s+=list[i]

mean=s/len(list)

print("Mean of the list is : ",mean)

list.sort()
if len(list)%2==0:
    median=list[len(list)/2]
else:
    median=(list[(len(list)//2)-1]+list[len(list)//2])/2

print("median of the given list is : ",median)

mode = 3*(mean-median)-1
print("mode of the given list is : ",mode)

variance=0

for i in range(len(list)):
    variance+=(pow((list[i]-mean),2))

variance=variance/(len(list)-1)

print("Variance of the given list is : ",variance)


SD=pow(variance,0.5)
print("Standard Deviation of the given list is : ",SD)
```

### AP formula:
*   a2 = a + d
*   an = a + (n-1) * d
*   sn = (n/2) * (2*a + (n - 1) * d) 

```python
def AP(a,n,d):
    yield a
    c=a+d
    while n>0:
       yield c
       c+=d
       n-=1
       
print(list(AP(1,4,2)))
```

### GP formula:
*   a2 = a *  r
*   an = a * r**(n-1)
*   sn = (a * (1-r**n))/(1-r) 


```python
def GP(a,n,r):
    i=1
    while i<=n:
       out=a*r**(i-1)
       yield out
       i+=1
print(list(GP(1,4,2)))
```

```python
pi=(22/7)

class Circle:
    def __init__(self,r): ## constructor/init method
        self.r=r ## self id used to address an object

    def area(self):
        return (pi*self.r**2)
    def circum(self):
        return (2*pi*self.r)
    def diametre(self):
        return 2*self.r
    
c=Circle(7)
```
